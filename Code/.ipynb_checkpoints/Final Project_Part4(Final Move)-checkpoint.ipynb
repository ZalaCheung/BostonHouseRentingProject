```python
df = pd.read_csv('house_renting_dataset_Merge2.csv',index_col=0)
df = df.reset_index()
BasicDict = df.to_dict()
```

```python
priceCluster = []
crimeCluster = []
restaurantCluster = []
grocerieCluster = []
cafeCluster = []
shoppingCluster = []
nightlifeCluster = []
schoolCluster = []
combineData = []
needToScale = []
coordinateData = []
schoolData = []
priceData = []
crimeData = []

for i in range(len(BasicDict['Cafes'])):
    priceCluster.append([])
    crimeCluster.append([])
    restaurantCluster.append([])
    grocerieCluster.append([])
    cafeCluster.append([])
    shoppingCluster.append([])
    nightlifeCluster.append([])
    schoolCluster.append([])
    combineData.append([])
    needToScale.append([])
    coordinateData.append([])
    schoolData.append([])
    priceData.append([])
    crimeData.append([])
```

```python
for i in range(len(BasicDict['Cafes'])):
    # coordinate
    longitude = BasicDict['Longitude'][i]
    latitude = BasicDict['Latitude'][i]
    
    # 有数值的
    price = BasicDict['Price_per_room'][i]    
    cafeNumber = BasicDict['Cafes'][i]
    grocerieNumber = BasicDict['Groceries'][i]
    nightlifeNumber = BasicDict['Nightlife'][i]
    restaurantNumber = BasicDict['Restaurants'][i]
    schoolNumber = BasicDict['School'][i]
    shoppingNumber = BasicDict['Shopping'][i]
    
    #无数值的
    crimeDescribe = BasicDict['Crime'][i]
    if crimeDescribe == "Lowest":
        crimeRate = 0.1
    elif crimeDescribe == "Low" :
        crimeRate = 0.35
    elif crimeDescribe == "High":
        crimeRate = 0.65
    elif crimeDescribe == "Highest":
        crimeRate = 0.9
    else:
        print("something wrong about the crime rate!")
        
    highSchoolDescribe = BasicDict['High_School'][i]
    if highSchoolDescribe == "Above Average":
        highSchoolRate = 0.8
    elif highSchoolDescribe == "Average":
        highSchoolRate = 0.5
    elif highSchoolDescribe == "Below Average":
        highSchoolRate = 0.2
    else:
        print("something wrong about the highSchool rate!" + str(i))
        
    middleSchoolDescribe = BasicDict['Middle_School'][i]
    if middleSchoolDescribe == "Above Average":
        middleSchoolRate = 0.8
    elif middleSchoolDescribe == "Average":
        middleSchoolRate = 0.5
    elif middleSchoolDescribe == "Below Average":
        middleSchoolRate = 0.2
    else:
        print("something wrong about the middleSchool rate!")
    
    schoolRate = (highSchoolRate + 0.8 * middleSchoolRate) /2
        
    combineData[i] = [longitude, latitude, price, restaurantNumber, 
                      grocerieNumber, cafeNumber, shoppingNumber, 
                      nightlifeNumber, schoolRate, crimeRate]
    coordinateData[i] = [longitude, latitude]
    priceData[i] = [price]
    schoolData[i] = [(highSchoolRate + 0.8 * middleSchoolRate) / 2]
    crimeData[i] = [crimeRate]
    needToScale[i] = [restaurantNumber, grocerieNumber, cafeNumber, 
                      shoppingNumber, nightlifeNumber]
```

```python
coordinateArray = np.array(coordinateData)
priceArray = np.array(priceData)
schoolArray = np.array(schoolData)
crimeArray = np.array(crimeData)
```

```python
# Error detect
error = np.zeros(20)
for k in range(1,20):
    kmeans = KMeans(init='k-means++', n_clusters=k, n_init=10)
    kmeans.fit_predict(matrix)
    error[k] = kmeans.inertia_
plt.plot(range(1,len(error)),error[1:])
plt.xlabel('Number of clusters')
plt.title('K-means1')
dummy = plt.ylabel('Error')
```

```python
# S score
def sc_evaluate_clusters(X,max_clusters):
    s = np.zeros(max_clusters+1)
    s[0] = 0;
    s[1] = 0;
    for k in range(2,max_clusters+1):
        kmeans = KMeans(init='k-means++', n_clusters=k, n_init=10)
        kmeans.fit_predict(X)
        s[k] = metrics.silhouette_score(X,kmeans.labels_,metric='cosine')
    plt.plot(range(2,len(s)),s[2:])
    plt.xlabel('Number of clusters')
    plt.ylabel('Adjusted Rand Index')
    plt.title('K-means2')
sc_evaluate_clusters(standardInputData,20)
```

```python
'''
K-Means Method
'''
kmeans = KMeans(init='k-means++', n_clusters= 12, n_init=300)
kmeans.fit_predict(standardInputData)
centers = kmeans.cluster_centers_
```

```python
'''
Use this function to get the index of cluster
'''
def Clusts(clustnum,labels_array):
    return np.where(labels_array == clustnum)[0]
```

```python
'''Use this function to get the labels of clusters'''

def get_labels(ids, testSet):
    labels = []
    for i in range(len(ids)):
        price_sum = 0
        restaurantNumber_sum = 0
        grocerieNumber_sum = 0
        cafeNumber_sum = 0
        shoppingNumber_sum = 0
        nightlifeNumber_sum = 0
        schoolRate_sum = 0
        crimeRate_sum = 0
        
        for j in range(len(ids[i])):
            price_sum += priceArray[ids[i][j]][0]
            restaurantNumber_sum += matrix[ids[i][j]][3] 
            grocerieNumber_sum += matrix[ids[i][j]][4]
            cafeNumber_sum += matrix[ids[i][j]][5]
            shoppingNumber_sum += matrix[ids[i][j]][6]                                   
            nightlifeNumber_sum += matrix[ids[i][j]][7]
            schoolRate_sum += matrix[ids[i][j]][8]
            crimeRate_sum += matrix[ids[i][j]][9]
                                               
            price_estimated = ("%.2f" %(price_sum / len(ids[i])))
            restaurant_estimated = ("%.0f" %(restaurantNumber_sum / len(ids[i])))
            grocerie_estimated = ("%.0f" %(grocerieNumber_sum / len(ids[i])))                                   
            cafe_estimated = ("%.0f" %(cafeNumber_sum / len(ids[i])))
            shopping_estimated = ("%.0f" %(shoppingNumber_sum / len(ids[i])))
            nightlife_estimated = ("%.0f" %(nightlifeNumber_sum / len(ids[i]))) 
            school_estimated = ("%.2f" %(schoolRate_sum / len(ids[i])))                                    
            crime_estimated = ("%.2f" %(crimeRate_sum / len(ids[i])))
        
        if testSet == "price":
            labels.append(price_estimated)
        elif testSet == "restaurant":
            labels.append(restaurant_estimated)
        elif testSet == "grocerie":
            labels.append(grocerie_estimated)
        elif testSet == "cafe":
            labels.append(cafe_estimated)
        elif testSet == "shopping":
            labels.append(shopping_estimated)
        elif testSet == "nightlife":
            labels.append(nightlife_estimated)
        elif testSet == "school":
            labels.append(school_estimated)
        elif testSet == "crime":
            labels.append(crime_estimated)
                                               
        #labels.append([price_estimated,restaurant_estimated,grocerie_estimated,
                       #cafe_estimated,shopping_estimated,nightlife_estimated,
                       #school_estimated,crime_estimated])
        #labels.append(cafeScore / len(ids[i]))
        #labels.append(shoppingScore / len(ids[i]))
    return labels
```

```python
'''Kmeans labels'''
ids_kmeans = []
for i in range(len(kmeans.cluster_centers_)):
    ids_kmeans.append(Clusts(i,kmeans.labels_))
price_label_kmeans = get_labels(ids_kmeans, "price")
restaurant_label_kmeans = get_labels(ids_kmeans, "restaurant")
grocerie_label_kmeans = get_labels(ids_kmeans, "grocerie")
cafe_label_kmeans = get_labels(ids_kmeans, "cafe")
shopping_label_kmeans = get_labels(ids_kmeans, "shopping")
nightlife_label_kmeans = get_labels(ids_kmeans, "nightlife")
school_label_kmeans = get_labels(ids_kmeans, "school")
crime_label_kmeans = get_labels(ids_kmeans, "crime")
```

```python
def getColors(num_colors):
    colors = []
    for i in np.arange(0., 360., 360. / num_colors):
        hue = i/360.
        lightness = (50 + np.random.rand() * 10) / 100.
        saturation = (90 + np.random.rand() *  10) / 100.
        colors.append(colorsys.hls_to_rgb(hue, lightness, saturation))
    return colors
```

```python
'''
GMM Method
'''
gmm = mixture.GMM(n_components=12, covariance_type='spherical')
gmm.fit(standardInputData)
```

```python
pro_m= gmm.predict_proba(standardInputData)
gmm_clusters_dic = {}
for i in range(np.matrix(pro_m).shape[0]):
    mat_list = np.matrix(pro_m)[i,:].tolist()[0]
#     print(type(mat_list))
    a = max(mat_list)
    for j in range(len(mat_list)):
        if a == mat_list[j]:
            max_index = j
    if str(max_index) not in gmm_clusters_dic:
        gmm_clusters_dic[str(max_index)]=[]
        gmm_clusters_dic[str(max_index)].append(i)
    gmm_clusters_dic[str(max_index)].append(i)
```

```python
'''GMM labels'''
ids_gmm = []
for k,v in gmm_clusters_dic.items():
    
    ids_gmm.append(gmm_clusters_dic[k])
label_gmm = get_labels(ids_gmm, "price")
```
